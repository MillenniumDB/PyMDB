```python
from pymdb import BatchLoader, MDBClient
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def draw_batch(batch):
    G = nx.Graph()
    for i in range(len(batch.node_features)):
        G.add_node(i)
    G.add_edges_from(batch.edge_index)
    plt.figure(figsize=(16, 9))
    nx.draw_networkx(G, with_labels=True)
    plt.show()
```

```python
with MDBClient(host="127.0.0.1", port=8080) as client:
    batch_loader = BatchLoader(
        client,
        feature_store_name="random",
        num_seeds=32,
        batch_size=8,
        neighbor_sizes=(5, 5),
        seed=2023,
    )

    for batch in batch_loader:
        print(batch)
        draw_batch(batch)
```
